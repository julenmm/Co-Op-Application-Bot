# Nu Works Application Bot

This is a Selenium bot that allowes you to apply to any job position on Simplicity (Application used by Nu Works)

## Features
- Simple login (add data in txt file)
- manual filtering of roles ( mannually apply filters before "stop running cells")
- Automatic application to All filtered possition
- Automatic resume selection
- Automatic Transcript selection
- Automatic Generic cover letter Selection
- Application log creation in Xcell File at the end of the program listing sucessful and failed applications separately


## How to Use

1. Install Jupyter notebooks and all requred libraries with pip install
- (requestes, BeautifulSoup, selenium)
2. Create a text file named "credentials.txt" in the same directory where this ipynb file is located
- In the file: 
    * Add your Username as a string as (username = 'my username')
    * Add your Password as a string as (password = 'my password')
    * Log in from a normal tab and in duo security verification click "send codes", enter codes as a list (access_codes = *****). Codes have to be formatted as strings
4. Enter your files in Nu Works.
    * Upload your resume and name it "Resume"
    * Upload a your transcript and name it "Transcript"
    * Upload a generic cover letter usable for as many possitions as possible and name it "Cover Letter"
    - IT IS IMPORTANT THAT THE NAMES ARE EXACTLY THE SAME.
3. Congrats!!! Now you're ready to start runnning the program, apply to absolutely EVERY position you filter, and stop wasting your time.
* Run cells by clicking shift + enter until you reach the cell that says "Stop Running"
* select Your desired filters for the jobs you want to apply (manually)
* Continue running using shift + enter

4. LET THE PROGRAM RUN AND APPLY FOR YOU OVER NIGHT

5. See the list of the applications you made and couldn't make as excel files in the same directory of this project.

6. Get Interview Requests like You're a celebrity and enjoy!!!!!!!!!!!!

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
import time
```

```python
# Define the file path
file_path = 'credentials.txt'

# Initialize variables to store username, password, and access codes
username = None
password = None
access_codes = []

# Open the file and read the lines
with open(file_path, 'r') as file:
    for line in file:
        # Split the line into key and value
        key, value = line.strip().split(': ')
        if key == 'username':
            username = value
        elif key == 'password':
            password = value
        elif key == 'access_codes':
            # Split the access codes into a list
            access_codes = value.split(', ')

# Print the extracted values
print(f'Username: {username}')
print(f'Password: {password}')
print(f'Access Codes: {access_codes}')
```

```python
class Application:
    def __init__(self, position_title, nu_works_url, to_apply_url, company_name):
        self.position_title = position_title
        self.nu_works_url = nu_works_url
        self.to_apply_url = to_apply_url
        self.company_name = company_name
```
